```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# plt.rcParams['figure.facecolor'] = 'w'
plt.style.use('dark_background')
import seaborn as sns
import random
```

```python
def f(x):
    return np.exp(-(x-5)**2/2)

def q(x):
    return 1.

def q2(x):
    return 1.3*np.ones(len(x))
```

```python
x_axis = np.linspace(0, 10, 100)
plt.plot(x_axis, f(x_axis));
plt.plot(x_axis, q2(x_axis));
```

$$ \lambda = \sup_x \frac{f(x)}{q(x)} $$

```python
np.random.seed(42)
k = 1.3
prop_samples = np.random.uniform(low=0.0, high=10, size=10000)
samples_accepted = []
for x0 in prop_samples:
    u = np.random.uniform(low=0.0, high=1, size=1)
    r = f(x0) / k #(k*q(x0))
    if u <= r:
        samples_accepted.append(x0)
    # print(xs, u)
len(prop_samples), len(samples_accepted), len(samples_accepted)/len(prop_samples)
```

```python
fig, axs = plt.subplots(figsize=(10, 9), nrows=2)
xmin, xmax = -0.5, 10.5
# ymin, ymax = 0, 0.5

ax = axs[0]
x_axis = np.linspace(0, 10, 1000)
ax.plot(x_axis, f(x_axis)*0.41, label='distribution')
ax.set(ylabel='pdf')

ax = axs[1]
ax.hist(samples_accepted, density=True, bins=30, histtype='step', label='sample distribution')
ax.set(ylabel='sample counts')
        
for ax in axs.flat:
    ax.legend()
    ax.set(xlabel='x', ylabel='sample counts', xlim=[xmin,xmax], )#ylim=[ymin,ymax])
    ax.grid(lw=1, alpha=0.2, zorder=0)
    
# fig.savefig('../plots/pen-draw-distributions-0-4.pdf', bbox_inches='tight')
```
